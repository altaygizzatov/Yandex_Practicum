# Описание проекта: Телекоммуникации

Оператор связи «ТелеДом» хочет бороться с оттоком клиентов. Для этого его сотрудники начнут предлагать промокоды и специальные условия всем, кто планирует отказаться от услуг связи. Чтобы заранее находить таких пользователей, «ТелеДому» нужна модель, которая будет предсказывать, разорвёт ли абонент договор. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и услугах. Ваша задача — обучить на этих данных модель для прогноза оттока клиентов.

**Описание услуг**

Оператор предоставляет два основных типа услуг: 
- Стационарную телефонную связь. Телефон можно подключить к нескольким линиям одновременно.
- Интернет. Подключение может быть двух типов: через телефонную линию (DSL, от англ. digital subscriber line — «цифровая абонентская линия») или оптоволоконный кабель (Fiber optic).

Также доступны такие услуги:
- Интернет-безопасность: антивирус (DeviceProtection) и блокировка небезопасных сайтов (OnlineSecurity);
- Выделенная линия технической поддержки (TechSupport);
- Облачное хранилище файлов для резервного копирования данных (OnlineBackup);
- Стриминговое телевидение (StreamingTV) и каталог фильмов (StreamingMovies).

Клиенты могут платить за услуги каждый месяц или заключить договор на 1–2 года. Возможно оплатить счёт разными способами, а также получить электронный чек.

**Описание данных.**

Данные состоят из нескольких файлов, полученных из разных источников:
- contract_new.csv — информация о договоре;
- personal_new.csv — персональные данные клиента;
- internet_new.csv — информация об интернет-услугах;
- phone_new.csv — информация об услугах телефонии.

Файл contract_new.csv:
- customerID — идентификатор абонента;
- BeginDate — дата начала действия договора;
- EndDate — дата окончания действия договора;
- Type — тип оплаты: раз в год-два или ежемесячно;
- PaperlessBilling — электронный расчётный лист;
- PaymentMethod — тип платежа;
- MonthlyCharges — расходы за месяц;
- TotalCharges — общие расходы абонента.

Файл internet_new.csv
- customerID — идентификатор пользователя;
- InternetService — тип подключения;
- OnlineSecurity — блокировка опасных сайтов;
- OnlineBackup — облачное хранилище файлов для резервного копирования данных;
- DeviceProtection — антивирус;
- TechSupport — выделенная линия технической поддержки;
- StreamingTV — стриминговое телевидение;
- StreamingMovies — каталог фильмов.

Файл phone_new.csv
- customerID — идентификатор пользователя;
- MultipleLines — подключение телефона к нескольким линиям одновременно.

Во всех файлах столбец customerID содержит код клиента. Информация о договорах актуальна на 1 февраля 2020 года.

## Исследовательский анализ и предобработка данных 

Вначале переведем название столбцов в змейиный регистр

```python
contract.columns = ['begin_date', 'end_date', 'type', 'paperless_billing', 'payment_method',
       'monthly_charges', 'total_charges']
contract.index.name = 'customer_id'
personal.columns = ['gender', 'senior_citizen', 'partner', 'dependents']
internet.columns =['internet_service', 'online_security', 'online_backup', 'device_protection',
       'tech_support', 'streaming_tv', 'streaming_movies']
phone.columns = ['multiple_lines']

for table in [contract, personal, internet, phone]:
    table.index.name = 'customer_id'
```

### contract

Переведем столбец в `total_charges` в числовой тип

```python
contract['total_charges'] = pd.to_numeric(contract['total_charges'], errors='coerce')
contract = contract.dropna(subset=['total_charges'])
```

Переведем все строковые столбцы в нижний регистр

```python
for table in [contract, personal, internet, phone]:
    for col in table.select_dtypes(include='object').columns:
        table[col] = table[col].apply(lambda x: x.lower())
```

Добавим новый столбец в `contract`, который показывает заканчивается ли контракт. И обработаем `end_date` вместо `no` вставим дату 1 февраля 2020, поскольку полученные данные актуальны на этот период и переведем столбец `end_date` в datetime 

```python
contract['will_end'] = contract['end_date'].apply(lambda x: 'no' if x=='no' else 'yes')
contract['end_date'] = contract['end_date'].apply(lambda x: '2020-02-01' if x=='no' else x)
contract['end_date'] = pd.to_datetime(contract['end_date'])
```

Удалим явные дубликаты из таблицы `contract`, поскольку они будут мешать обучению моделей

```python
contract['contract_duration_days'] = (contract['end_date'] - contract['begin_date']).dt.days
contract = contract.drop(['end_date', 'begin_date'], axis=1)
```

```python
def pieplot(data):
    plt.figure(figsize=(12, 10))
    plt.title('Соотношение категориальных признаков')

    j = 0
    columns = data.columns.to_list()
    for i in range(len(columns)):
        j += 1
        try:
            pivot_table = data.pivot_table(index=columns[i], values=columns[j], aggfunc={'count'})
        except:
            pivot_table = data.pivot_table(index=columns[i], values=columns[i-1], aggfunc={'count'})
        ax = plt.subplot(2, len(columns)//2, j)
        ax.pie(pivot_table['count'], labels=pivot_table.index, autopct='%.0f%%')
        ax.set_title(columns[i])

    plt.tight_layout();
```

## Подготовка данных

Разделим столбцы с целевой переменной и признаками, и провдем разделение выборок

```python
def data_splitter(data, columns):
    encoder = LabelEncoder()
    target = encoder.fit_transform(data['will_end'])
    features = data[columns].drop('will_end', axis=1)

    ohe_columns = features.select_dtypes(include='object').columns.to_list()
    num_columns = features.select_dtypes(include='number').columns.to_list()


    features_train, features_test, target_train, target_test = train_test_split(features,
                                                                                target,
                                                                                test_size=TEST_SIZE,
                                                                                stratify=target,
                                                                                random_state=RANDOM_STATE)
    print(f'Кодировка целевой переменной 0 и 1 {encoder.classes_} соответственно')
    
    return features_train, features_test, target_train, target_test, ohe_columns, num_columns
```

Для LogisticRegression

```python
features_train, features_test, target_train, target_test, ohe_columns, num_columns = data_splitter(final, columns)
```

Для Модели RandomForestClassifier, CatBoostClassifier

```python
(features_train_forest, 
 features_test_forest, 
 target_train_forest, 
 target_test_forest, 
 ohe_columns_forest, 
 num_columns_forest) = data_splitter(final, final.columns)
```

Подготовим пайплайн для кодировки признаков

```python
def data_preprocessor_pipe(ohe_columns, num_columns):
    ohe_pipe = Pipeline(
        [
            (
                'simpleImputer_ohe', 
                SimpleImputer(missing_values=np.nan, strategy='most_frequent')
            ),
            (
                'ohe', 
                OneHotEncoder(drop='first', handle_unknown='ignore')
            )
        ]
    ) 
    data_preprocessor = ColumnTransformer(
        [
            ('ohe', ohe_pipe, ohe_columns),
            ('num', MinMaxScaler(), num_columns)
        ], 
        remainder='passthrough'
    )
    return data_preprocessor
```

## Обучение моделей

### LogisticRegression

```python
linear_pipe = Pipeline([
    ('preprocessor', data_preprocessor_pipe(ohe_columns, num_columns)),
    ('models', LogisticRegression(random_state=RANDOM_STATE))
])

param_grid = [
    {
        'models__C': range(1, 10),
        'models__solver': ['lbfgs', 'liblinear', 'saga']
    }
]

grid_search = GridSearchCV(
    linear_pipe, 
    param_grid, 
    cv=5,
    scoring='roc_auc',
    n_jobs=-1
)

grid_search.fit(features_train, target_train)
```

```python
print('\nЛучшая модель и её параметры:\n\n', grid_search.best_estimator_)
print('Метрика ROC-AUC для лучшей модели:\n', grid_search.best_score_)
```

### RandomForestClassifier

```python
forest_pipe = Pipeline([
    ('preprocessor', data_preprocessor_pipe(ohe_columns_forest, num_columns_forest)),
    ('models', RandomForestClassifier(random_state=RANDOM_STATE))
])

param_grid = {
    'models__max_depth': [None] + [i for i in range(2, 7)],
    'models__min_samples_split': range(2, 10)
             }

randomized_search = RandomizedSearchCV(
    forest_pipe, 
    param_grid, 
    cv=3,
    scoring='roc_auc',
    random_state=RANDOM_STATE,
    n_jobs=-1
)

randomized_search.fit(features_train_forest, target_train_forest)
```

```python
print('\nЛучшая модель и её параметры:\n\n', randomized_search.best_estimator_)
print('Метрика ROC-AUC для лучшей модели:\n', randomized_search.best_score_)
```

### CatBoostClassifier

```python
cat_pipe = Pipeline([
    ('models', CatBoostClassifier(random_state=RANDOM_STATE, 
                              loss_function='Logloss',
                              logging_level='Silent',
                              cat_features=ohe_columns_forest
                             ))
])

param_grid = {
    'models__depth':[None] + [i for i in range(1, 10)],
    'models__learning_rate': np.arange(0.1, 1.0, 0.3),
    'models__n_estimators': np.arange(100, 1500, 200)
}


catboost_search = RandomizedSearchCV(cat_pipe, 
                                     param_grid, 
                                     cv=3,
                                     scoring='roc_auc',
                                     random_state=RANDOM_STATE,
                                     n_jobs=-1
                                    )

catboost_search.fit(features_train_forest, target_train_forest)
```

```python
print('\nЛучшая модель и её параметры:\n', catboost_search.best_params_)
print('Метрика ROC-AUC для лучшей модели:\n', catboost_search.best_score_)
```

```python
param = catboost_search.best_params_
param['n_estimators'] = param.pop('models__n_estimators')
param['learning_rate'] = param.pop('models__learning_rate')
param['depth'] = param.pop('models__depth')
param['random_state'] = RANDOM_STATE
param['loss_function'] = 'Logloss'
param['logging_level'] = 'Silent'


catboost = CatBoostClassifier(**param)
catboost.fit(Pool(features_train_forest, target_train_forest, cat_features=ohe_columns_forest))
```

- Лучше всех себя показала CatBoostClassifier

Поскольку нам нужно предсказать расторжение контракта ошибки FP критичны, рассчитаем метрику Precision и попробуем минимизировать ее, изменением порога значимости

```python
data = pd.DataFrame(zip(target_test_forest, target_proba),
             columns = ['y_test', 'y_proba']).sort_values(by='y_proba',ascending=False)

thresholds = [round(i,2) for i in np.linspace(0.5, 1, num = 15, endpoint=False)]

columns = []
thresholds_values = []
fp = []
precision = []

conf_matrix = pd.DataFrame()

for i in thresholds:
    columns.append('y_pred_'+str(i))
    data['y_pred_'+str(i)] = data['y_proba'].apply(lambda x: 1 if x>=i  else 0)
    thresholds_values.append(i)
    fp.append(confusion_matrix(data['y_test'], data['y_pred_'+str(i)])[0][1])
    precision.append(precision_score(target_test_forest, np.where(target_proba <= i, 0, 1)))

conf_matrix['thresholds_values'] = thresholds_values
conf_matrix['FP'] = fp
conf_matrix['Precision'] = precision
plt.figure(figsize=(8,4))
sns.barplot(data=conf_matrix, x='thresholds_values', y='Precision')
plt.ylim(0.85, 1.0)
plt.title('Зависимость Precision от порога значимости');
```

Как видно из графика взяв пороговое значени 0.67 мы увеличим Precision, но FP значения все равно будут ненулевые

```python
threshold = 0.67
```

```python
cm = confusion_matrix(target_test_forest, np.where(target_proba <= threshold, 0, 1))
ax = sns.heatmap(cm, cmap='coolwarm', annot=True, fmt='.0f')
plt.show()
```

FP ошибки снизилась до 5

Посмотрим на Accuracy с этим порогом

```python
catboost_accuracy = accuracy_score(target_test_forest, np.where(target_proba <= threshold, 0, 1))
catboost_precision = precision_score(target_test_forest, np.where(target_proba <= threshold, 0, 1))
print(f'Accuracy {catboost_accuracy}, при threshold {threshold}')
print(f'Precision {catboost_precision}, при threshold {threshold}')
```

Она не сильно упала, значит все хорошо
